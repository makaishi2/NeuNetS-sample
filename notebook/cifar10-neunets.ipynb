```python
from keras.datasets import cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print(x_train.shape, y_train.shape)
```

```python
# 作業用ディレクトリの準備
base_dir = 'Image'
zip_dir = 'ZIP'

import os
os.mkdir(base_dir)
os.mkdir(zip_dir)
```

```python
import numpy as np
cifar10_labels = np.array([
    'airplane',
    'automobile',
    'bird',
    'cat',
    'deer',
    'dog',
    'frog',
    'horse',
    'ship',
    'truck'])
```

```python
for i in range(10):
    subdir = '%s/%s' % (base_dir, cifar10_labels[i])
    print(subdir)
    os.mkdir(subdir)
```

```python
base_dir = 'Image'
from PIL import Image

label_file = 'labels.csv'
f = open(label_file, "w")

# 生成イメージ数(最大50000)
image_num = 1000

for i in range(image_num):
    img = Image.fromarray(x_train[i])
    label_i = y_train[i][0]
    label = cifar10_labels[label_i]
    #filename = '%s/%s/train-%s-%05d.png' % (base_dir, label, label, i)
    filename = '%s/%s/train-%s-%05d.jpg' % (base_dir, label, label, i)
    print(filename)
    f.write('%s,%s\n' % (filename, label))
    with open(filename, mode='wb') as out:
        img.save(out, format='jpeg', quality=100, optimize=True)
        #img.save(out, format='png')
f.close()
```
